To perform the transformation of points in Parts 1-4 induced by matrix A, we can define the coordinates of each point in each part as lists and store them in separate lists for each part. We can then apply the matrix transformation to each point in each part by multiplying the point's coordinates by the matrix A.

```python
# Importing libraries
import numpy as np

def exercise_1a(a, b):

    # Define the coordinates of each point in each part
    part1 = [[a, a], [a, -a], [-a, -a], [-a, a], [a, a]]
    part2 = [[1 - a, a - 1], [-a/4, a - 1], [-a/4, a/2], [1 - a, a/2], [1 - a, a - 1]]
    part3 = [[a/4, a - 1], [a - 1, a/2], [a - 1, a - 1], [a/4, a/2]]
    part4 = [[a - 1, -a/4], [a - 1, -a/2], [1 - a, -a/2], [1 - a, -a/4]]

    # Define the matrix A
    transformation_matrix = np.array([[1, 1/b], [-1/b, 1/b]])

    # Empty list that will contain transformed parts coordinates
    transformed_parts = []

    # Function to transform a part
    def transform_part(part):
        # Initialising a list that will contain transformed coordinates
        transformed_part = []
        # Iterating trought each point
        for point in part:
            # Transforming the point
            transformed_point = np.dot(transformation_matrix, point).tolist()
            # Adding it to the list of transformed points
            transformed_part.append(transformed_point)

        # adding transformed part to the list with all transformed parts
        transformed_parts.append(transformed_part)

        return transformed_part

    # Transforming each part
    transformed_part1 = transform_part(part1)
    transformed_part2 = transform_part(part2)
    transformed_part3 = transform_part(part3)
    transformed_part4 = transform_part(part4)
    
    # Printing out the coordinates
    print("Transformed coordinates of Part 1: ", transformed_part1)
    print("Transformed coordinates of Part 2: ", transformed_part2)
    print("Transformed coordinates of Part 3: ", transformed_part3)
    print("Transformed coordinates of Part 4: ", transformed_part4)

    # returning all transformed parts so they can be used in the next exercise
    return transformed_parts

# calling the function and defining a and b as parameters
exercise_1a(4, 3)
```

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def exercise_1b(a, b):

    part1 = [[a, a], [a, -a], [-a, -a], [-a, a], [a, a]]
    part2 = [[1 - a, a - 1], [-a / 4, a - 1], [-a / 4, a / 2], [1 - a, a / 2], [1 - a, a - 1]]
    part3 = [[a / 4, a - 1], [a - 1, a / 2], [a - 1, a - 1], [a / 4, a / 2]]
    part4 = [[a - 1, -a / 4], [a - 1, -a / 2], [1 - a, -a / 2], [1 - a, -a / 4]]

    transformed_parts = exercise_1a(a, b)

    transformed_part1 = transformed_parts[0]
    transformed_part2 = transformed_parts[1]
    transformed_part3 = transformed_parts[2]
    transformed_part4 = transformed_parts[3]

    def plot_parts(pt1, pt2, pt3, pt4, colour):

        x_values = [[pt1[0][0], pt1[1][0]], [pt1[1][0], pt1[2][0]], [pt1[2][0], pt1[3][0]],
                    [pt1[3][0], pt1[4][0]]]

        y_values = [[pt1[0][1], pt1[1][1]], [pt1[1][1], pt1[2][1]], [pt1[2][1], pt1[3][1]],
                    [pt1[3][1], pt1[4][1]]]

        x_values_pt2 = [[pt2[0][0], pt2[1][0]], [pt2[1][0], pt2[2][0]], [pt2[2][0], pt2[3][0]],
                        [pt2[3][0], pt2[4][0]]]
        y_values_pt2 = [[pt2[0][1], pt2[1][1]], [pt2[1][1], pt2[2][1]], [pt2[2][1], pt2[3][1]],
                        [pt2[3][1], pt2[4][1]]]

        x_values_pt3 = pt3[0][0], pt3[1][0]
        y_values_pt3 = pt3[0][1], pt3[1][1]

        x_values_pt3_1 = pt3[2][0], pt3[3][0]
        y_values_pt3_1 = pt3[2][1], pt3[3][1]

        x_values_pt4 = [[pt4[0][0], pt4[1][0]], [pt4[1][0], pt4[2][0]], [pt4[2][0], pt4[3][0]]]
        y_values_pt4 = [[pt4[0][1], pt4[1][1]], [pt4[1][1], pt4[2][1]], [pt4[2][1], pt4[3][1]]]

        plt.plot(x_values, y_values, colour)
        plt.plot(x_values_pt2, y_values_pt2, colour)
        plt.plot(x_values_pt3, y_values_pt3, colour)
        plt.plot(x_values_pt3_1, y_values_pt3_1, colour)
        plt.plot(x_values_pt4, y_values_pt4, colour)

    plt.xlabel('x')
    plt.ylabel('y')
    plot_parts(part1, part2, part3, part4, 'red')
    red_patch = mpatches.Patch(color='red', label='Before transformation')
    plot_parts(transformed_part1, transformed_part2, transformed_part3, transformed_part4, 'blue')
    blue_patch = mpatches.Patch(color='blue', label='After transformation')
    plt.legend(handles=[red_patch, blue_patch], bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower right", mode="expand",
               borderaxespad=0, ncol=3)
    plt.show()

# calling the function and defining a and b as parameters
exercise_1b(4, 3)
```

To get the exact solution I used the the integration factor method as follows:

Step 1: Write the differential equation in the form of dy/dx + P(x)y = Q(x).
dy/dx = x + y/5 can be written as dy/dx - y/5 = x.

Step 2: Identify the values of P(x) and Q(x).
Here, P(x) = -1/5 and Q(x) = x.

Step 3: Find the integration factor, which is e^(P(x)dx).
P(x)dx = (-1/5)dx = -x/5
Therefore, the integration factor is e^(-x/5).

Step 4: Multiply both sides of the differential equation by the integration factor.
e^(-x/5)dy/dx - (1/5)e^(-x/5)y = xe^(-x/5)

Step 5: Apply the product rule to the left side and simplify.
(d/dx)(e^(-x/5)y) = xe^(-x/5)
Integrating both sides, we get
e^(-x/5)y = xe^(-x/5)dx
e^(-x/5)y = -5xe^(-x/5) - 5e^(-x/5)dx + C
where C is an arbitrary constant of integration.

Simplifying the integral on the right side, we get
e^(-x/5)y = -5xe^(-x/5) - 5e^(-x/5) + C
Multiplying both sides by e^(x/5), we get
y = -5x - 25 + Ce^(x/5)

Step 6: Use the initial condition y(0) = -3 to solve for C.
y(0) = -5(0) - 25 + Ce^(0/5) = -3
C = 22

Therefore, the exact solution is:
y = -5x - 25 + 22e^(x/5)

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, x0, y0, h, xn):
    n = int((xn - x0) / h)  # number of steps
    x = [x0]  # x values
    y = [y0]  # y values

    for i in range(n):
        y.append(y[i] + h * f(x[i], y[i]))
        x.append(x[0] + (i+1)*h)
    return x, y


# define the derivative function
def f(x, y):
    return x + y/5


# set initial conditions and step sizes
x0 = 0
y0 = -3
h1 = 1
h2 = 0.2
h3 = 0.05
xn = 5


# define the exact solution function
def exact_solution(x):
    return -5*x - 25 + 22*np.exp(x / 5)
    

# define the range and the number of solutions
x_exact = np.linspace(x0, xn, 1000)

# apply Euler's method with different step sizes
x1, y1 = euler_method(f, x0, y0, h1, xn)
x2, y2 = euler_method(f, x0, y0, h2, xn)
x3, y3 = euler_method(f, x0, y0, h3, xn)

# plot the numerical solutions
plt.plot(x1, y1, label='h=1')
plt.plot(x2, y2, label='h=0.2')
plt.plot(x3, y3, label='h=0.05')

# plot the exact solution
plt.plot(x_exact, exact_solution(x_exact), label='exact')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

In conclusion I found out the the smaller the step size is, the closer solution is to the exact solution. Given that Euler's method doesn't require to solve the differrential equation, with the small enough step size it is more useful as it is not limited to the complexity of the differential equation and on average has a better Big O.
